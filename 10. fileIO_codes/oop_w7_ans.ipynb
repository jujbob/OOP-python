### 1. 

$x = [1, 2, 3, ..., 100]$ 에 대하여, "square.txt" 파일의 각 줄에 
x의 제곱수를 저장하는 코드를 작성하세요.

ex.



````python
output_file = open("square.txt", 'w')

def square(x):
    return x**2

for i in range(1, 101):
    output_file.writelines(f"{square(i)}\n")
    
output_file.close()
````

### 2.

1번 문제에서 저장한 "square.txt"파일을 읽어온 뒤, 각 줄에 있는 값을 int 형태로 순서대로 list에 저장하세요.

````python

input_file = open("square.txt", 'r')
arr = []

while True:
    input_str = input_file.readline()
    
    if input_str == "":
        break
    
    arr.append(int(input_str))

input_file.close()
````

### 3. 
제공된 no3_data.txt 파일의 각 줄을 읽어서, 해당 줄에 있는 텍스트를 car, fruit, number로 분류하여 새로운 파일 no3_result.txt로 저장하세요.

ex.



````python
car = ["truck", "bus", "k3", "bmw", "benz"]
fruit = ["orange", "apple", "grape"]
number = ["1", "1234", "124", "1240124"]

with open("no3_data.txt", "r", encoding="utf-8") as file:
    for line in file.readlines():
        if line[:-1] in car:
            line = line.replace(line[:-1], "car")
        elif line[:-1] in fruit:
            line = line.replace(line[:-1], "fruit")
        else:
            line = line.replace(line[:-1], "number")
    
        with open("./target.txt", "a", encoding="utf-8") as write_file:
            write_file.writelines(line)
````

### 4. with open 사용하기

open() 함수를 사용할 때, close() 함수를 통해 수동으로 닫아줘야 한다. 
with open()을 사용하게 되면, with open 구문이 종료될 때 자동으로 파일이 닫힌다.

ex.

```py
with open("test.txt", 'w') as f:
    f.writelines("test")
```

### 5. "append" 모드 

파일 write 시, `'w'` 인자를 주게 되는데, 이 동작은 기존에 파일이 있으면 "덮어쓰기"를 하게 된다.

`'a'` 인자를 사용하면 기존에 존재하는 파일 가장 뒤에 write하고자 하는 텍스트를 저장하게 된다.

with open과 append 모드를 사용해서 "n_lines.txt" 파일에 다음을 저장하는 코드를 작성하세요.

1 번째 줄입니다.

2 번째 줄입니다.

...

10 번째 줄입니다.

````python
for i in range(1, 11):
    with open("n_lines.txt", 'a') as f:
        f.writelines(f"{i} 번째 줄입니다.\n")
````
